```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv("Salary.csv")
df
```

```python
#scatterplot
%matplotlib inline
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='green',marker='*')
```

```python
new_df= df.drop('Salary',axis='columns')
new_df.head()
```

```python
Salary= df.Salary
Salary
```

```python
#create linear regression object
reg=linear_model.LinearRegression()
reg.fit(new_df,Salary)
```

```python
reg.predict([[34]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
34* 8731.94104706 + 28859.52619453144
```

```python
reg.predict([[10]])
```

```python
YOE_df=pd.read_csv('YOE.csv')
YOE_df.head()
```

```python
p=reg.predict(YOE_df)
p
```

```python
YOE_df['Salary']=p
YOE_df
```

```python
YOE_df.to_csv("Predicted Salary.csv")
```
